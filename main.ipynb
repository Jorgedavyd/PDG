```python
from geopy.distance import geodesic

import copy
def dataframe_distance(presence, independent, down_boundary: int, up_boundary: int):

    presence_points = presence.loc[:, ['Latitude', 'Longitude']].values
    
    filtered_dataframe = copy.deepcopy(independent)
    


    for idx, point_1 in enumerate(presence_points):
        filtered_dataframe[f'distance_{idx+1}'] = geodesic(point_1, (filtered_dataframe['Latitude'], filtered_dataframe['Longitude']))

    for point_1 in presence_points:
        for idx_2, point_2 in enumerate(filtered_dataframe.loc[:, ['Latitude', 'Longitude']].values):
            if down_boundary<geodesic(point_1, point_2).kilometers<up_boundary:
                pass
            else:
                filtered_dataframe = filtered_dataframe.drop(idx_2, axis = 0)
    
    return filtered_dataframe
```
